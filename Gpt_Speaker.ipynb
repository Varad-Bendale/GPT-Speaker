CNN is a tool clearly used to reduce or rather normalise the huge sound magnitude data to shorter ones which hepns us make it better later on . the clear mechanism is to simply multiply these magntudes with random set of terms normalising factor and add them and make them shorter

```python

class CNN() :
  def __init__(  num_of_things_for_mult , num_of_things  , normalised_magnitudes) :
     self.num_of_things_for_mult = num_of_things_for_mult
     self.num_of_things = num_of_things
     self.normalised_magnitudes = normalised_magnitudes
     self.stride = 2

  def generate_random_num(self):
     std   = sqrt(2.0 / self.num_of_things)
     random_numbers = []
     for i in range(self.num_of_things_for_mult) :
         random_numbers.append(randn(0, std))
     return random_numbers

  def compression(self, normalised_magnitudes):
      numbers = self.generate_random_num()
      magni = []
      for i in range(0, len(normalised_magnitudes) - self.num_of_things_for_mult + 1, stride):
        cnn_num = 0
        temp_mags = self.normalised_magnitudes[i : i + self.num_of_things_for_mult ]
        for j in range(len(temp_mags)) :
            cnn_num += (temp_mags[j] * numbers[j] )
        magni.append(cnn_num)
      return magni

  def relu(self, magni):
    output = []
    for x in magni:
        if x > 0:
            output.append(x)
        else:
            output.append(0)
    return output

  def forward(self, normalised_magnitudes):
    x = self.compression(normalised_magnitudes)
    x = self.relu(x)
    return x
```

```python
from sklearn.decomposition import PCA
import torch.nn as nn

class VQVAE(nn.Module) :
  def __init__(embeddings ,normalised_magnitudes ):
       self.embeddings = embeddings
       self.codebook = nn.Parameter(torch.randn(512, 32))
       self.normalised_magnitudes = normalised_magnitudes

  def encoder(self) :
     layer1 = CNN(num_of_things_for_mult=3, num_of_things=32)
     layer2 = CNN(num_of_things_for_mult=3, num_of_things=16)
     embeddings = []
     for i in range(len(normalised_magnitudes)):
        x = layer1.forward(normalised_magnitudes[i])
        x = layer2.forward(x)
        embeddings.append(x.flatten())
     return embeddings

  def visualize(self):
    pca = PCA(n_components=2)
    codebook_2d = pca.fit_transform(self.codebook.detach().numpy())
    plt.scatter(codebook_2d[:, 0], codebook_2d[:, 1])
    plt.show()

  def closest_term(self , embeds ) :
      distances = torch.cdist(embeds, self.codebook)
      nearest_index = torch.argmin(distances, dim=1)
      nearest_term = self.codebook[nearest_index]
      return nearest_index , nearest_term

  def training(self) :
    info = {}
    for k in range(50) :
        if k > 0 and k % 10 == 0 :
           new_vocab(self.embeddings , self.codebook )
        details = {}

        for i in range(len(self.embeddings)) :
            nearest_index , codeword  = closest_term(self , self.embeddings[i] )
            ind = nearest_index.item()
            details.setdefault(ind, []).append(self.embeddings[i])

        with torch.no_grad():
            for ind in details:
                average = torch.stack(details[ind]).mean(dim=0)
                self.codebook[ind] = 0.99 * self.codebook[ind] + 0.01 * average
                info[ind] = {"embeddings": details[ind],"codeword": self.codebook[ind]}
    return info

   def new_vocab(self.embeddings , self.codebook ) :
       distances = torch.cdist(embeddings, self.codebook)
       topk = torch.topk(distances, k=3, largest=False)
       codewords = self.codebook[topk.indices]
       weights = 1 / (topk.values + 1e-8)
       weights = weights / weights.sum(dim=1, keepdim=True)
       count  = {}
       for i in range(len(self.embeddings)) :
           info = (weights[i], codewords[i])
           current_ind = tuple(topk.indices[i].tolist())
           if info[0][0] < 0.5  :
              new_vector = info[1][0]*info[0][0]+ info[1][1]*info[0][1]+ info[1][2]*info[0][2]
              if len(self.codebook) > 0 :
                 distances = torch.cdist(new_vector.unsqueeze(0), self.codebook)
                 if distances.min() < 0.3 :
                    nearest_index = torch.argmin(distances, dim=1).item()
                    count[nearest_index] = count.get(nearest_index, 0) + 1
                    continue
              if current_ind not in self.count:
                 self.count[current_ind] = 0
              else  :
                 count[current_ind] = count.get(current_ind, 0) + 1
              if count[current_ind] == 30 :
                 new_vector = info[1][0]*info[0][0]+ info[1][1]*info[0][1]+ info[1][2]*info[0][2]
                 self.codebook = nn.Parameter(torch.cat([self.codebook, new_vector.unsqueeze(0)], dim=0))



   def rvq(self) :
        embeddings_tensor = torch.stack(self.embeddings)
        nearest_index, nearest_terms = self.closest_term(embeddings_tensor)
        rvq = embeddings_tensor - nearest_terms
       return rvq



   def get_tokens(self):
       info  = training()
       token_ids = list(info.keys())
       return torch.tensor(token_ids)

```
