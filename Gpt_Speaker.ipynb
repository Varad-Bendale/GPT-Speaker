CNN is a tool clearly used to reduce or rather normalise the huge sound magnitude data to shorter ones which hepns us make it better later on . the clear mechanism is to simply multiply these magntudes with random set of terms normalising factor and add them and make them shorter

```python

class CNN() :
  def __init__(  num_of_things_for_mult , num_of_things  , normalised_magnitudes) :
     self.num_of_things_for_mult = num_of_things_for_mult
     self.num_of_things = num_of_things
     self.normalised_magnitudes = normalised_magnitudes
     self.stride = 2

  def generate_random_num(self):
     std   = sqrt(2.0 / self.num_of_things)
     random_numbers = []
     for i in range(self.num_of_things_for_mult) :
         random_numbers.append(randn(0, std))
     return random_numbers

  def compression(self, normalised_magnitudes):
      numbers = self.generate_random_num()
      magni = []
      for i in range(0, len(normalised_magnitudes) - self.num_of_things_for_mult + 1, stride):
        cnn_num = 0
        temp_mags = self.normalised_magnitudes[i : i + self.num_of_things_for_mult ]
        for j in range(len(temp_mags)) :
            cnn_num += (temp_mags[j] * numbers[j] )
        magni.append(cnn_num)
      return magni

  def relu(self, magni):
    output = []
    for x in magni:
        if x > 0:
            output.append(x)
        else:
            output.append(0)
    return output

  def forward(self, normalised_magnitudes):
    x = self.compression(normalised_magnitudes)
    x = self.relu(x)
    return x
```

```python
class Head(nn.Module):
    def __init__(self, head_quant):
        super().__init__()
        self.query = nn.Linear(dim, head_quant, bias=False)
        self.key   = nn.Linear(dim, head_quant, bias=False)
        self.value = nn.Linear(dim, head_quant, bias=False)
        self.mask  = torch.tril(torch.ones(block_size, block_size))

    def forward(self, x):
        q = self.query(x)
        k = self.key(x)
        v = self.value(x)
        T = x.shape[1]
        wieghts = q @ k.transpose(-2,-1) * (head_quant ** -0.5)
        wieghts = wieghts.masked_fill(self.mask[:T, :T].to(x.device) == 0, float('-inf'))
        wieghts = torch.softmax(wieghts, dim=-1)
        pos_pattern = wieghts @ v
        return pos_pattern

class head_concat(nn.Module):
    def __init__(self, num_heads, head_quant):
        super().__init__()
        self.heads = nn.ModuleList([Head(head_quant) for _ in range(num_heads)])
        self.proj  = nn.Linear(num_heads * head_quant, dim)

    def forward(self, x):
        pos_pattern = torch.cat([head(x) for head in self.heads], dim=-1)
        return self.proj(pos_pattern)

class block_in_transformer(nn.Module) :

    def __init__(self) :
         super().__init__()
         self.head = head_concat(num_heads, head_quant)
         self.hidden_layer_1 = nn.Linear(dim , 4*dim).to(device)
         self.hidden_layer_2  = nn.Linear(4*dim , dim).to(device)
         self.norm_1 = nn.LayerNorm(dim)
         self.norm_2 = nn.LayerNorm(dim)


    def forward(self , x) :
       pattern = self.head(x)
       x = x + pattern
       x  = self.norm_1(x)
       ffn = torch.nn.functional.leaky_relu(self.hidden_layer_1(x), negative_slope=0.01)
       ffn = self.hidden_layer_2(ffn)
       x = x + ffn
       x = self.norm_2(x)
       return x
```

```python
import torch.nn as nn
from google.colab import files

class personality_transformer(nn.Module):

      def __init__(self  , clutch ) :
          super().__init__()
          self.device = 'cuda' if torch.cuda.is_available() else 'cpu'
          self.n_layers   = 4
          self.num_heads  = 4
          self.head_quant = 64
          self.batch_size = 64
          self.Blocks = nn.ModuleList([block_in_transformer() for _ in range(self.n_layers)]).to(self.device)
          self.g = torch.Generator()
          self.emotion = {'01': 'neutral','02': 'calm','03': 'happy','04': 'sad','05': 'angry','06': 'fearful','07': 'disgust','08': 'surprised'}
          self.g = self.g.manual_seed(42)

          if clutch == 0 :
              self.dim = 256
              self.block_size = 10
              self.positions  = nn.Embedding(self.block_size, self.dim).to(self.device)
              self.info = face_data
              self.vocab_size = codebook_face_vqvae_data
              self.Encods = nn.Embedding(self.vocab_size , self.dim).to(self.device)
          if clutch == 1 :
              self.dim = 80
              self.block_size = 64
              self.positions  = nn.Embedding(self.block_size, self.dim).to(self.device)
              self.info = sound_data
              self.vocab_size = vqvae_sound_codebook_size
              self.Encods = nn.Embedding(self.vocab_size , self.dim).to(self.device)

          self.final = nn.Linear(self.dim, self.vocab_size).to(self.device)
          self.x_part = []
          self.x_tokens   = []
          self.y_tokens  = []
          self.xs = []
          self.ys  = []

       def make_data(self, dataset):
          self.xs = []
          self.ys = []
          self.x_tokens = torch.tensor(dataset[0])
          self.y_tokens = torch.tensor(dataset[1])

       def defining_part(self):
            self.x_tokens = torch.stack(self.x_tokens)
            self.y_tokens = torch.stack(self.y_tokens)

            for i in range(0, len(self.x_tokens) - self.block_size):
                self.xs.append(self.x_tokens[i : i + self.block_size])
                self.ys.append(self.y_tokens[i : i + self.block_size])

            self.xs = torch.stack(self.xs)
            self.ys = torch.stack(self.ys)

      def training(self) :
          self.parameters = list(self.Encods.parameters()) + list(self.positions.parameters()) + list(self.Blocks.parameters()) + list(self.final.parameters())
          self.optimizer = torch.optim.AdamW(self.parameters, lr=1e-4)
          for i in range(10000):
            ix  = torch.randint(0, self.xs.shape[0], (self.batch_size,))
            x_  = self.xs[ix].to(self.device)
            y_  = self.ys[ix].to(self.device)
            tok_emb = self.Encods(x_)
            size = x_.shape[1]
            pos_emb = self.positions(torch.arange(size).to(self.device))
            x = tok_emb + pos_emb
            for block in self.Blocks:
                x = block(x)
            logits = self.final(x)
            loss = torch.nn.functional.cross_entropy(logits.view(-1, self.vocab_size),y_.view(-1))
            self.optimizer.zero_grad()
            loss.backward()
            self.optimizer.step()
            if i % 500 == 0:
                print(f'step {i} loss: {loss.item():.4f}')
            if loss.item() < 1.3:
                print(f'stopping at step {i}')
                break


      def decode(self, sound):
          vqvae.get_tokens(sound)
          x_ = torch.tensor(self.x_part[-1]).unsqueeze(0).to(self.device)
          tok_emb = self.Encods(x_)
          size = x_.shape[1]
          pos_emb = self.positions(torch.arange(size).to(self.device))
          x = tok_emb + pos_emb
          for block in self.Blocks:
              x = block(x)
          logits  = self.final(x)
          predicted_word = torch.argmax(logits, dim=-1)
          return predicted_word


        def save(self, emotion_num):
            emotion_name = self.emotion[emotion_num]
            if clutch == 1:
                name = f'sound_personality_{emotion_name}.pt'
            if clutch == 0:
                name = f'face_personality_{emotion_name}.pt'
            torch.save({'Encods': self.Encods.state_dict(), 'positions': self.positions.state_dict(), 'blocks': self.Blocks.state_dict(), 'final': self.final.state_dict()}, name)
            files.download(name)
```
