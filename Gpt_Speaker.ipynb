```python
# just for fetching the frequencies of the data
import librosa

audio, sr = librosa.load("speech.wav", sr=16000)
frequencies = librosa.stft(audio)
```

```python
import math
magnitudes = []
for freq in frequencies :
    magnitudes.append((freq.real**2 + freq.imag**2)** 0.5 )
```

```python
max_mag = 1e+8
min_mag = 1e-8
normalised_magnitudes = []
for mag in magnitudes :
    max_mag = max(max_mag  , mag)
    min_mag = max(min_mag  , mag)
for freq in frequencies :
    normalised_magnitudes.append((freq - min_freq)/(max_freq - min_freq))
```

```python

class CNN() :
  def __init__(  num_of_things_for_mult , num_of_things  , normalised_magnitudes) :
     self.num_of_things_for_mult = num_of_things_for_mult
     self.num_of_things = num_of_things
     self.normalised_magnitudes = normalised_magnitudes
     self.stride = 2

  def generate_random_num(self):
     std   = sqrt(2.0 / self.num_of_things)
     random_numbers = []
     for i in range(self.num_of_things_for_mult) :
         random_numbers.append(randn(0, std))
     return random_numbers

  def compression(self, normalised_magnitudes):
      numbers = self.generate_random_num()
      magni = []
      for i in range(0, len(normalised_magnitudes) - self.num_of_things_for_mult + 1, stride):
        cnn_num = 0
        temp_mags = self.normalised_magnitudes[i : i + self.num_of_things_for_mult ]
        for j in range(len(temp_mags)) :
            cnn_num += (temp_mags[j] * numbers[j] )
        magni.append(cnn_num)
      return magni

  def relu(self, magni):
    output = []
    for x in magni:
        if x > 0:
            output.append(x)
        else:
            output.append(0)
    return output

  def forward(self, normalised_magnitudes):
    x = self.compression(normalised_magnitudes)
    x = self.relu(x)
    return x



layer1 = CNN(num_of_things_for_mult=3, num_of_things=256 , normalised_magnitudes)
layer2 = CNN(num_of_things_for_mult=3, num_of_things=128, normalised_magnitudes)
layer3 = CNN(num_of_things_for_mult=3, num_of_things=64, normalised_magnitudes)

x = layer1.forward(magnitudes)
x = layer2.forward(x)
x = layer3.forward(x)
```
